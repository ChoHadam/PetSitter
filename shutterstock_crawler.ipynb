```python
from selenium import webdriver  # 라이브러리(모듈) 가져오라
from selenium.webdriver import ActionChains as AC
import requests
import urllib.request
from bs4 import BeautifulSoup
import re
from time import sleep
import time
```

```python
driver = webdriver.Chrome("chromedriver.exe")  # 크롬 드라이버 로드
# driver.get("https://www.shutterstock.com/ko/search/sick+dogs?image_type=photo&page=1")  # "#차량용방향제" 검색
# time.sleep(5)
```

```python
searchword="dog+angry"

for n in range(1,10000000):
    url=f"https://www.shutterstock.com/ko/search/{searchword}?image_type=photo&page={n}"
    driver.get(url)
    page=driver.find_element_by_css_selector("input[aria-label='Change page']")
    
    if n!=int(page.get_attribute("value")):
        break
        
    results=driver.find_elements_by_css_selector("img.z_h_c.z_h_e")
    for i,result in enumerate(results):
        urllib.request.urlretrieve(result.get_attribute("src"),f"angry/{20*(n-1)+i}.jpg")
    time.sleep(10)
```
