네이버 이미지 크롤링 코드

```
from bs4 import BeautifulSoup 
import requests
import json
import urllib.request

# keyword : 검색 키워드,  n : 크롤링할 데이터 수(20의 배수로 넣어야함)
def naver_img_crawler():
  count=0
  keyword = input('찾고자하는 검색 키워드를 입력하세요. : ')
  
  isInt = False
  while isInt == False: 
    n = int(input('크롤링할 데이터 수 n을(n >= 100) 50의 배수로 넣으세요.(예: 100, 150, 1000) : '))
    if n % 50 == 0 and n >= 100:
      isInt  = True

  #처음 50장 이미지 추출
  res = requests.get('https://m.search.naver.com/search.naver?sm=mtb_hty.top&where=m_image&oquery=&tqi=Uwu5Slp0JWVsskpQPyRssssstDZ-354428&query={}'.format(keyword))
  bs = BeautifulSoup(res.text, 'html.parser')

  for i in bs.select('body div#container section.sc.sp_nimage._prs_img div._box a img'):
    count += 1 
    imgUrl = i.attrs['data-source']
    urllib.request.urlretrieve(imgUrl, keyword.replace(' ', '_') + str(count))
    print(keyword.replace(' ', '_') + str(count), imgUrl)
    
  #51장 부터 100장씩 추출 (비동기 크롤링 방식)
  start = 51
  while start < n:
    res2 = requests.get('https://s.search.naver.com/imagesearch/instant/search.naver?where=image&section=image&rev=42&res_fr=0&res_to=0&face=0&color=0&ccl=0&ac=0&aq=0&spq=0&query={}&nx_search_query={}&nx_and_query=&nx_sub_query=&nx_search_hlquery=&nx_search_fasquery=&datetype=0&startdate=0&enddate=0&json_type=6&nlu_query=&nqx_theme=%7B%22theme%22%3A%7B%22main%22%3A%7B%22name%22%3A%22dictionary_en%22%7D%2C%22sub%22%3A%5B%7B%22name%22%3A%22language%22%7D%5D%7D%7D&start={}&display=50'.format(keyword, keyword, start))
    refined_text = res2.text[3:-4]
    res2_dics = json.loads(refined_text)

    for i in res2_dics['result']['items']['item']:
      count += 1
      bs2 = BeautifulSoup(i['html'],'html.parser')
      imgUrl = bs2.select_one('div a img').attrs['data-source']
      urllib.request.urlretrieve(imgUrl, keyword.replace(' ', '_') + str(count))
      print(keyword.replace(' ', '_') + str(count), imgUrl)
    
    start += 50

#https://s.search.naver.com/imagesearch/instant/search.naver?where=image&section=image&rev=42&res_fr=0&res_to=0&face=0&color=0&ccl=0&ac=0&aq=0&spq=0&query=dog&nx_search_query=dog&nx_and_query=&nx_sub_query=&nx_search_hlquery=&nx_search_fasquery=&datetype=0&startdate=0&enddate=0&json_type=6&nlu_query=&nqx_theme=%7B%22theme%22%3A%7B%22main%22%3A%7B%22name%22%3A%22dictionary_en%22%7D%2C%22sub%22%3A%5B%7B%22name%22%3A%22language%22%7D%5D%7D%7D&start=151&display=100&_callback=window.__jindo2_callback.__sauImageTabList_0
```
