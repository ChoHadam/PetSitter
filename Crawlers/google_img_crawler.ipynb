구글 이미지 크롤링 코드

```
from bs4 import BeautifulSoup 
import requests
import urllib.request

# keyword : 검색 키워드,  n : 크롤링할 데이터 수(20의 배수로 넣어야함)
def goodle_img_crawler():
  count = 0

  keyword = input('찾고자하는 검색 키워드를 입력하세요. : ')
  isInt = False
  while isInt == False: 
    n = int(input('크롤링할 데이터 수 n을(n >= 20) 20의 배수로 넣으세요.(예: 20, 40, 1000) : '))
    if n % 20 == 0 and n >= 20 :
      isInt  = True
  n = int(n/20)

  for i in range(0,n):
    link = 'https://www.google.com/search?q={}&hl=ko&biw=365&bih=844&tbm=isch&ei=vHn9XujHCrKSr7wP7PK_yAg&start={}&sa=N'.format(keyword,i*20 )
    res = requests.get(link)
    bs = BeautifulSoup(res.text, 'html.parser')
    
    for i in bs.select('body table.GpQGbf tr td.e3goi table.TxbwNb a div.RAyV4b img'):
      count += 1
      imgUrl = i.attrs['src']
      urllib.request.urlretrieve(imgUrl, keyword.replace(' ', '_') + str(count))
      print(keyword.replace(' ', '_') + str(count), imgUrl)
```

크롤링 함수 호출

```
goodle_img_crawler()
```
